```python
import pandas as pd
import numpy as np
```

```python
#Retrieve all records from the Olympics table.
df = pd.read_csv("Olympics.csv")
df
```

```python
#Retrieve only the Athlete names and the Medals they won
athelete_medals = df[['name','medal']]
athelete_medals
```

```python
#. Count the total number of athletes in the dataset
total_atheletes = df['name'].nunique()
total_atheletes
```

```python
#. Find all records of athletes who won a Gold medal
gold_medal_atheletes = df[df['medal'] == 'Gold']
gold_medal_atheletes
```

```python
#List all athletes who won Silver, ordered by the Year they won it
silver_medal_atheletes = df[df['medal'] == 'Silver']
silver_medal_Order = silver_medal_atheletes.sort_values(by = 'year')
silver_medal_Order
```

```python
#Count how many Gold, Silver, and Bronze medals each country has won
medal_counts = df.groupby(['noc','medal']).size()
medal_counts
```

```python
#. Identify countries that have won more than 50 Gold medals.
gold_medal = df[df['medal'] == 'Gold']
gold_medal_counts = gold_medal.groupby('noc').size()
gold_medal_counts
```

```python
#Olympics data with country information to show the total number of medals won by countries with a population over 10 million. not possible becuase the dataset is not appropriate according to question
```

```python
#Find the athlete who has won the most medals
athlete_medal_counts = df.groupby('name')['medal'].count()
athlete_with_most_medals = athlete_medal_counts.idxmax()
athlete_with_most_medals
```

```python
#List all events that include the term 'Freestyle' in their name
freestyle_events = df[df['event'].str.contains('Freestyle', case=False)]
freestyle_events['event'].unique()
```

```python
#. Find the top 3 athletes by the total number of medals won in each sport.
athlete_medal_counts = df.groupby(['sport', 'name'])['medal'].count().reset_index()
athlete_rankings = athlete_medal_counts.sort_values(by=['medal'], ascending=False).head(3)
athlete_rankings
```

```python
#List athletes who won more than one medal in a single Olympic year
athlete_medal_counts = df.groupby(['year', 'name'])['medal'].count().reset_index()
multi_medal_winners = athlete_medal_counts[athlete_medal_counts['medal'] > 1].sort_values(by='medal',ascending=False)
multi_medal_winners
```

```python
#Identify countries that have won gold medals in both Summer and Winter Olympics (assuming a 'Season' column exists).
gold_medals = df[df['medal'] == 'Gold']
gold_medal_counts = gold_medals.groupby(['noc', 'season']).size().unstack(fill_value=0)
countries_with_gold_in_both_seasons = gold_medal_counts[(gold_medal_counts['Summer'] > 0) & (gold_medal_counts['Winter'] > 0)]
countries_with_gold_in_both_seasons
```

```python
#Calculate the average number of medals won per athlete for each country.
medal_counts_per_athlete = df.groupby(['noc', 'name'])['medal'].count().reset_index()
average_medals_per_athlete = medal_counts_per_athlete.groupby('noc')['medal'].mean()
average_medals_per_athlete
```

```python
# List countries that have won medals in more than 10 different sports.
countries_multiple_sports = df.groupby('noc')['sport'].nunique()
countries_more_than_10_sports = countries_multiple_sports[countries_multiple_sports > 10].index.tolist()
print(countries_more_than_10_sports)
```

```python
# Show the year difference between the first and last medal won by each country.

first_last_medal_years = df.groupby('noc')['year'].agg(['min', 'max'])
first_last_medal_years['Year_Difference'] = first_last_medal_years['max'] - first_last_medal_years['min']
print(first_last_medal_years[['Year_Difference']])
```
